## 2. Carga de archivos PDF

## 3. Exploracion del segundo documento (2018-2022)

Antes de consolidar, identificamos qué páginas del segundo documento
contienen la tabla equivalente de STEAM por sexo y año.

## 5. Construccion de DataFrames por documento

```python
# DataFrame Documento 1: 2014-2021
años_doc1 = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

df_doc1 = pd.DataFrame({
    'año': años_doc1,
    'total_diplomas': [14990, 15470, 15501, 16793, 16944, 17347, 16051, 21303],
    'steam_total':    [5169,  5525,  5711,  6468,  6154,  6654,  5830,  7007],
    'steam_mujer':    [2259,  2445,  2553,  2836,  2816,  3049,  2828,  3246],
    'steam_hombre':   [2910,  3080,  3158,  3632,  3338,  3605,  3002,  3761],
    'no_steam_total': [9821,  9945,  9790,  10325, 10790, 10693, 10221, 14296],
    'no_steam_mujer': [6583,  6778,  6586,  7172,  7535,  7510,  7287,  10223],
    'no_steam_hombre':[3238,  3167,  3204,  3153,  3255,  3183,  2934,  4073],
    'fuente': 'OPES_2014_2021'
})

print("DataFrame Documento 1 creado ✓")
print(df_doc1)
```

```python
# DataFrame Documento 2: 2018-2022
# Nota: los datos de 2018-2021 servirán para validación cruzada
# El dato nuevo que aporta este documento es 2022
años_doc2 = [2018, 2019, 2020, 2021, 2022]

df_doc2 = pd.DataFrame({
    'año': años_doc2,
    'total_diplomas': [16944, 17347, 16051, 21303, 20754],
    'steam_total':    [6154,  6654,  5830,  7007,  7273],
    'steam_mujer':    [2816,  3049,  2828,  3246,  3440],
    'steam_hombre':   [3338,  3605,  3002,  3761,  3833],
    'no_steam_total': [10790, 10693, 10221, 14296, 13481],
    'no_steam_mujer': [7535,  7510,  7287,  10223, 9646],
    'no_steam_hombre':[3255,  3183,  2934,  4073,  3835],
    'fuente': 'OPES_2018_2022'
})

print("DataFrame Documento 2 creado ✓")
print(df_doc2)
```

## 6. Validación cruzada — periodo de solapamiento 2018–2021

Ambos documentos contienen datos para los años 2018, 2019, 2020 y 2021.
Verificamos que los números coincidan exactamente entre ambas fuentes.
Una discrepancia indicaría un error de transcripción o una revisión
posterior de los datos por parte de OPES-CONARE.

```python
# Validación cruzada en el periodo de solapamiento 2018-2021
años_solapamiento = [2018, 2019, 2020, 2021]

doc1_solap = df_doc1[df_doc1['año'].isin(años_solapamiento)].reset_index(drop=True)
doc2_solap = df_doc2[df_doc2['año'].isin(años_solapamiento)].reset_index(drop=True)

columnas_validar = ['año', 'total_diplomas', 'steam_total', 'steam_mujer', 'steam_hombre']

print("=== DOCUMENTO 1 (2014-2021) — años 2018-2021 ===")
print(doc1_solap[columnas_validar].to_string(index=False))

print("\n=== DOCUMENTO 2 (2018-2022) — años 2018-2021 ===")
print(doc2_solap[columnas_validar].to_string(index=False))

print("\n=== RESULTADO DE VALIDACIÓN CRUZADA ===")
for col in ['total_diplomas', 'steam_total', 'steam_mujer', 'steam_hombre']:
    coincide = (doc1_solap[col].values == doc2_solap[col].values).all()
    print(f"{col}: {'✓ Coincide' if coincide else '✗ Discrepancia'}")
```

## 7. Consolidacion del dataset final 2014-2022

```python
# Consolidación: tomamos doc1 para 2014-2021 y agregamos 2022 desde doc2
df_2022 = df_doc2[df_doc2['año'] == 2022].copy()

df_final = pd.concat([df_doc1, df_2022], ignore_index=True)

# Columna adicional: porcentaje STEAM del total
df_final['pct_steam'] = (df_final['steam_total'] / df_final['total_diplomas'] * 100).round(1)

# Columna adicional: porcentaje mujeres dentro de STEAM
df_final['pct_steam_mujer'] = (df_final['steam_mujer'] / df_final['steam_total'] * 100).round(1)

# Columna adicional: porcentaje hombres dentro de STEAM
df_final['pct_steam_hombre'] = (df_final['steam_hombre'] / df_final['steam_total'] * 100).round(1)

print("=== DATASET FINAL CONSOLIDADO 2014-2022 ===")
print(f"Periodo: {df_final['año'].min()} – {df_final['año'].max()}")
print(f"Registros: {len(df_final)}")
print()
print(df_final[['año', 'total_diplomas', 'steam_total', 'pct_steam',
                'steam_mujer', 'pct_steam_mujer',
                'steam_hombre', 'pct_steam_hombre']].to_string(index=False))
```

```python
# Exportar dataset final a CSV

df_final.to_csv("graduados_steam_consolidado_2014_2022.csv", index=False)

print("Dataset exportado: graduados_steam_consolidado_2014_2022.csv ✓")
print(f"Registros: {len(df_final)}")
print(f"Columnas: {df_final.columns}")
```
