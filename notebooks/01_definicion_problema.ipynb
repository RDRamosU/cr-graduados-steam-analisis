## 1. Contexto

Costa Rica ha apostado históricamente por la educación superior como motor
de desarrollo económico. En los últimos años, la demanda de profesionales
en áreas STEAM (Science, Technology, Engineering, Arts and Mathematics)
ha crecido significativamente, impulsada por la llegada de empresas
multinacionales de tecnología y el crecimiento del sector de servicios.

Sin embargo, existe poca visibilidad pública sobre cuántos profesionales
STEAM produce el sistema universitario estatal costarricense, en qué áreas
se concentran y si existe una brecha de género en estas disciplinas.

Este proyecto analiza los datos oficiales de diplomas otorgados por las
cinco universidades estatales de Costa Rica (UCR, TEC, UNA, UNED, UTN)
durante el periodo 2014–2022, con foco específico en la clasificación STEAM.

## 2. Pregunta central

> por año y en qué áreas se concentran?

## 3. Preguntas de análisis
