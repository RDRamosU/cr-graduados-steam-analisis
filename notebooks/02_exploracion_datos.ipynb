## 3. Carga de los archivos PDF

```python
import requests

# Archivo 1: periodo 2014-2021
url_1 = "https://github.com/RDRamosU/cr-graduados-steam-analisis/raw/main/data/raw/OPES_44_2023_Picado_C_estadisticas_diplomas_universidades_estatales_2014_2021.pdf"

response_1 = requests.get(url_1)
with open("OPES_2014_2021.pdf", "wb") as f:
    f.write(response_1.content)

# Archivo 2: periodo 2018-2022
url_2 = "https://github.com/RDRamosU/cr-graduados-steam-analisis/raw/main/data/raw/OPES_108_2023_Picado_C_estadisticas_diplomas_otorgados_universidades_estatales_2018_2022.pdf"

response_2 = requests.get(url_2)
with open("OPES_2018_2022.pdf", "wb") as f:
    f.write(response_2.content)

print(f"Archivo 1: {len(response_1.content):,} bytes ✓")
print(f"Archivo 2: {len(response_2.content):,} bytes ✓")
```

## 4. Extracción de tabla principal — STEAM por sexo y año (2014–2021)

Fuente: Página 25 del documento OPES 2014–2021  
Tabla: *Cantidad de diplomas otorgados en las universidades estatales
por clasificación STEAM y sexo, según año. 2014–2021*

```python
# Extracción del texto de la página 25

with pdfplumber.open("OPES_2014_2021.pdf") as pdf:
    texto_p25 = pdf.pages[24].extract_text()

print(texto_p25)
```

## 5. Construccion del DataFrame -STEAM por sexo y año

```python
# Construcción manual del DataFrame a partir del texto extraído
# Fuente: Tabla página 25 — OPES 2014-2021

años = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

data_steam_sexo = {
    'año': años,
    'total_diplomas': [14990, 15470, 15501, 16793, 16944, 17347, 16051, 21303],
    'steam_total':    [5169,  5525,  5711,  6468,  6154,  6654,  5830,  7007],
    'steam_mujer':    [2259,  2445,  2553,  2836,  2816,  3049,  2828,  3246],
    'steam_hombre':   [2910,  3080,  3158,  3632,  3338,  3605,  3002,  3761],
    'no_steam_total': [9821,  9945,  9790,  10325, 10790, 10693, 10221, 14296],
    'no_steam_mujer': [6583,  6778,  6586,  7172,  7535,  7510,  7287,  10223],
    'no_steam_hombre':[3238,  3167,  3204,  3153,  3255,  3183,  2934,  4073],
}

df_steam_sexo = pd.DataFrame(data_steam_sexo)

print("DataFrame creado exitosamente")
print(f"Filas: {df_steam_sexo.shape[0]} | Columnas: {df_steam_sexo.shape[1]}")
print("\nPrimeras filas:")
df_steam_sexo.head()
```

## 6. Inspeccion inicial del DataFrame

```python
print("=== INFORMACIÓN GENERAL ===")
print(df_steam_sexo.info())
print("\n=== ESTADÍSTICAS DESCRIPTIVAS ===")
df_steam_sexo.describe()
```

```python
# Verificación de consistencia: steam_total debe ser igual a steam_mujer + steam_hombre
df_steam_sexo['verificacion_steam'] = (
    df_steam_sexo['steam_total'] ==
    df_steam_sexo['steam_mujer'] + df_steam_sexo['steam_hombre']
)

print("=== VERIFICACIÓN DE CONSISTENCIA ===")
print(df_steam_sexo[['año', 'steam_total', 'steam_mujer', 'steam_hombre', 'verificacion_steam']])
print(f"\nTodos los registros son consistentes: {df_steam_sexo['verificacion_steam'].all()}")
```

## 7. Hallazgos preliminares de la exploración

A partir de la inspección inicial se identifican los siguientes patrones:

- **Tendencia general:** Los diplomas STEAM muestran crecimiento entre 2014 y 2021,
  con una caída notable en 2020 (5,830) posiblemente relacionada con la pandemia
  COVID-19, seguida de una recuperación fuerte en 2021 (7,007).

- **Brecha de género:** En todos los años los hombres superan a las mujeres en
  graduaciones STEAM, aunque la brecha se ha reducido progresivamente.

- **Salto 2020–2021:** El total de diplomas pasó de 16,051 en 2020 a 21,303 en 2021,
  un incremento del 32.8%. Este comportamiento atípico requiere análisis adicional
  en el Notebook 03.

- **Consistencia de datos:** Todos los registros pasan la verificación de consistencia
  (steam_total = steam_mujer + steam_hombre). ✓
